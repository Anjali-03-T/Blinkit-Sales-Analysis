```python
df["Quantity"] = df["Quantity"].replace("two", 2)
```

```python
df[["Quantity", "Unit_Price", "Total_Sales"]].head(10)
```

```python
df[df["Quantity"] * df["Unit_Price"] != df["Total_Sales"]]
```

```python
invalid_sales = df[
    (df["Quantity"].notna()) &
    (df["Unit_Price"].notna()) &
    (df["Total_Sales"].notna()) &
    (df["Quantity"] * df["Unit_Price"] != df["Total_Sales"])
]

invalid_sales
```

```python
len(invalid_sales)
```

```python
invalid_sales[["Quantity", "Unit_Price", "Total_Sales"]].head(10)
```

```python
df["Calculated_Quantity"] = df["Total_Sales"] / df["Unit_Price"]

df[["Quantity", "Calculated_Quantity"]].head(20)
```

```python
df_clean = df.copy()
```

```python
df_clean["Quantity"] = df_clean["Quantity"].fillna(
    df_clean["Total_Sales"] / df_clean["Unit_Price"]
)
```

```python
mask = (
    df_clean["Quantity"].notna()
    & df_clean["Unit_Price"].notna()
    & df_clean["Total_Sales"].notna()
)

df_clean.loc[mask, "Quantity"] = (
    df_clean.loc[mask, "Total_Sales"] / df_clean.loc[mask, "Unit_Price"]
)
```

```python
invalid = df_clean[
    (df_clean["Quantity"] * df_clean["Unit_Price"]) != df_clean["Total_Sales"]
]

len(invalid)
```

```python
df_clean.isnull().sum()
```

```python
df_clean["Unit_Price"] = df_clean["Unit_Price"].fillna(
    df_clean["Total_Sales"] / df_clean["Quantity"]
)
```

```python
df_clean["Total_Sales"] = df_clean["Total_Sales"].fillna(
    df_clean["Quantity"] * df_clean["Unit_Price"]
)
```

```python
df_clean.isnull().sum()
```

```python
df_clean[df_clean["Unit_Price"].isnull()]
```

```python
df_clean[df_clean["Unit_Price"].isnull()][["Quantity","Unit_Price","Total_Sales"]]
```

```python
df_clean[df_clean["Total_Sales"].isnull()][["Quantity","Unit_Price","Total_Sales"]]
```

```python
df_clean = df_clean.dropna(subset=["Unit_Price", "Total_Sales"])
```

```python
df_clean.isnull().sum()
```

```python
df_clean.drop(columns=["Calculated_Quantity"], inplace=True)
```

```python
df_clean["Date"].head(10)
```

```python
df_clean["Date"] = pd.to_datetime(
    df_clean["Date"],
    format="mixed",
    dayfirst=True
)
```
